```python
#
import pygame
import os
import math

import time
from random import random
import socket
import pickle
import threading
import struct
import numpy as np
from datetime import datetime

import random
def recvall(sock, n):
    # Helper function to recv n bytes or return None if EOF is hit
    data = bytearray()
    while len(data) < n:

        packet = sock.recv(n - len(data))
        if not packet:
            return None
        data.extend(packet)
    return data


def recv_msg(sock):
    # Read message length and unpack it into an integer
    raw_msglen = recvall(sock, 4)
    if not raw_msglen:
        return None
    msglen = struct.unpack('>I', raw_msglen)[0]
    #    Read the message data
    return recvall(sock, msglen)

#squadra=0
#x=random.randint(0,SIZECANVAS-1)
#y=random.randint(0,SIZECANVAS-1)
client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
#serverip = "2.tcp.eu.ngrok.io"
#port = 18682
code=input("server code: ")

serverip=code[0]+".tcp.eu.ngrok.io"
port=int(code[1:])
client.settimeout(None)

#
#
addr = (serverip, port)
client.connect(addr)


squadra=int(client.recv(1).decode("ascii"))
print(squadra)


client.send(b"1")

SIZECANVAS=int(client.recv(1).decode("ascii"))
client.send(b"1")
x=pickle.loads(recv_msg(client))
y=x[squadra][1]
x=x[squadra][0]
client.send(b"1")
timetomove=int(client.recv(1))
print(SIZECANVAS)
client.recv(1)

map=np.random.randint(1, size=(SIZECANVAS**2,SIZECANVAS**2))

#probability=np.random.rand(SIZECANVAS**2,SIZECANVAS**2)
#np.putmask(map, probability>connectionpercentage, 0)
team=np.ones(shape=(SIZECANVAS,SIZECANVAS),dtype=int)*-1
troups=np.zeros(shape=(SIZECANVAS,SIZECANVAS),dtype=int)


BLACK = (0, 0, 0)
WHITE = (200, 200, 200)
WINDOW_HEIGHT = 700
WINDOW_WIDTH = 600
connectionpercentage=0.2

#map=np.random.randint(SIZECANVAS**2, size=(SIZECANVAS,SIZECANVAS,SIZECANVAS**2))
#probability=np.random.rand(SIZECANVAS, SIZECANVAS,SIZECANVAS**2)
#np.putmask(map, probability>connectionpercentage, -1)
selected=np.zeros(shape=(SIZECANVAS,SIZECANVAS),dtype=bool)
#troups=np.zeros(shape=(SIZECANVAS,SIZECANVAS),dtype=int)
vision=np.zeros(shape=(SIZECANVAS,SIZECANVAS),dtype=int)



pygame.init()
SCREEN = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT), pygame.RESIZABLE)
CLOCK = pygame.time.Clock()
SCREEN.fill(BLACK)
blockxsize = WINDOW_WIDTH / SIZECANVAS
blockysize = (WINDOW_HEIGHT*6/7) / SIZECANVAS
def clamp(value):
    if value>=blockxsize/2:
        return SIZECANVAS/2
    if value<= -blockxsize/2:
        return -SIZECANVAS/2
    return value
SHIFTFACTOR=8

for a in range(SIZECANVAS**2):
    map[a][a]=0

selec=[]
cord=[]

#team=np.ones(shape=(SIZECANVAS,SIZECANVAS),dtype=int)*-1
colors=np.random.randint(0,255,size=(136,3),dtype=int)
colors[0][:]=[168, 79, 15]
colors[1][:]=[216, 0, 0]
colors[2][:]=[91, 110, 255]
colors[3][:]=[255, 94, 0]
colors[4][:]=[132, 64, 247]
colors[5][:]=[0, 213, 255]
colors[6][:]=[60, 213, 6]
colors[7][:]=[198, 13, 192]
colors[8][:]=[24, 0, 255]
colors[9][:]=[194, 120, 8]
colors[10][:]=[17, 166, 118]
colors[11][:]=[124, 6, 6]
colors[12][:]=[255, 255, 255]


clock=datetime.utcnow()
#print(x,y)
firstsecond=0

def connected(x,y,a,b):
    return map[y+x*SIZECANVAS][a+b*SIZECANVAS]
def neighvisible(x,y,d):

    for z in range(SIZECANVAS**2):
        if map[x*SIZECANVAS+y][z]:
            vision[z % SIZECANVAS ][z // SIZECANVAS ]+=1*d

def resource_path(relative_path):
    """ Get absolute path to resource, works for dev and for PyInstaller """
    try:
        # PyInstaller creates a temp folder and stores path in _MEIPASS
        base_path = sys._MEIPASS
    except Exception:
        base_path = os.path.abspath(".")

    return os.path.join(base_path, relative_path)
font = pygame.font.Font(resource_path("OpenSans-Bold.ttf"), 26)
def sendtroups(x,y,v):
    ms = pickle.dumps([1,(x,y),v])
    ms = struct.pack('>I', len(ms)) + ms
    client.sendall(ms)
def sendteam(x,y,v):
    ms = pickle.dumps([2,(x,y),v])
    ms = struct.pack('>I', len(ms)) + ms
    client.sendall(ms)
def sendmap(x,y,v):
    ms = pickle.dumps([3,(x,y),v])
    ms = struct.pack('>I', len(ms)) + ms
    client.sendall(ms)
tomove=[]
vision[x][y]=1
team[x][y]=squadra
troups[x][y]=1
neighvisible(x,y,1)
print(vision)
print(map)
xpos=WINDOW_WIDTH
sendteam(x,y,team[x][y])
sendtroups(x,y,troups[x][y])

timetoinc=10
tictac=datetime.utcnow()
destroy=0
colorsdestroy=[(35, 233, 180),(233, 80, 35)]

def handle(c,squadra):
    global troups,team,map
    while True:
        try:
            msg = recv_msg(c)
            if msg!=None:

                msg=pickle.loads(msg)
                print("RECEIVED")

                if msg[0]==1:
                    troups[msg[1][0]][msg[1][1]]=msg[2]
                elif msg[0]==2:
                    if team[msg[1][0]][msg[1][1]]==squadra and msg[2]!=squadra:
                        neighvisible(msg[1][0], msg[1][1], -1)
                        vision[msg[1][0]][msg[1][1]]-=1
                        selected[msg[1][0]][msg[1][1]]=0
                    team[msg[1][0]][msg[1][1]]=msg[2]

                elif msg[0]==3:
                    map[msg[1][0]][msg[1][1]]=msg[2]

        except:
            pass
receive_thread = threading.Thread(target=handle, args=(client,squadra))
receive_thread.start()

while True:

    if (datetime.utcnow()-tictac).total_seconds()>=timetoinc:
        tictac=datetime.utcnow()
        for x in range(SIZECANVAS):
            for y in range(SIZECANVAS):
                if team[x][y]==squadra:
                    troups[x][y]+=1
                    sendtroups(x,y,troups[x][y])

        #leggi dal server il permesso di incrementare
    SCREEN.fill(BLACK)

    # Set the size of the grid block3
    a, b = pygame.mouse.get_pos()
    if pygame.mouse.get_pressed()[0] and b>(WINDOW_HEIGHT*6/7):
        xpos=a

    rect = pygame.Rect(0,WINDOW_HEIGHT*6/7 ,
                       xpos,
                       WINDOW_HEIGHT*1/7 )
    pygame.draw.rect(SCREEN, colorsdestroy[destroy], rect)
    if firstsecond != 0:
        quantity=int(troups[(firstsecond-2)//SIZECANVAS][(firstsecond-2)%SIZECANVAS]*xpos/WINDOW_WIDTH+0.5)

        text = font.render(f'{quantity }', True, (0, 124, 98))
    else:
        text = font.render(f'SELECT', True, (0, 124, 98))
    textrect = text.get_rect()
    textrect.center = (WINDOW_WIDTH/2, WINDOW_HEIGHT*15/16)
    SCREEN.blit(text, textrect)

    for x in range(SIZECANVAS):
        for y in range(SIZECANVAS):
            if vision[x][y]:
                rect = pygame.Rect(x * blockxsize + blockxsize / 68, y * blockysize + blockysize / 68,
                                   blockxsize * 33 / 34,
                                   blockysize * 33 / 34)
                pygame.draw.rect(SCREEN, (35,38,180), rect)
                if selected[y][x]:
                    rect = pygame.Rect(x * blockxsize + blockxsize / 10, y * blockysize + blockysize / 10,
                                       blockxsize * 4 / 5,
                                       blockysize * 4 / 5)
                    pygame.draw.rect(SCREEN, (255, 138, 120), rect)
            else:
                rect = pygame.Rect(x * blockxsize, y * blockysize, blockxsize, blockysize)

                pygame.draw.rect(SCREEN, WHITE, rect, 1)


    for x in range(SIZECANVAS):
        for y in range(SIZECANVAS):
            if selected[y][x]:
                for z in range(SIZECANVAS**2):

                    if map[x*SIZECANVAS+y][z]:

                        if not selected[z//SIZECANVAS][z % SIZECANVAS]:
                            rect = pygame.Rect(z % SIZECANVAS * blockxsize + blockxsize / 10,
                                               z // SIZECANVAS * blockysize + blockysize / 10,
                                               blockxsize * 4 / 5,
                                               blockysize * 4 / 5)
                            pygame.draw.rect(SCREEN, (128, 238, 220), rect)
    for x in range(SIZECANVAS):
        for y in range(SIZECANVAS):
            if selected[y][x]:
                for z in range(SIZECANVAS ** 2):
                    if map[x*SIZECANVAS+y][z]:
                        pygame.draw.line(SCREEN,colors[x+y], (x * blockxsize+blockxsize/2+clamp((z//SIZECANVAS-SIZECANVAS/2)*SHIFTFACTOR/2), y * blockysize+blockysize/2+clamp((z%SIZECANVAS-SIZECANVAS/2)*SHIFTFACTOR/2)),(z%SIZECANVAS * blockxsize+blockxsize/2+clamp((z//SIZECANVAS-SIZECANVAS/2)*SHIFTFACTOR),z//SIZECANVAS  * blockysize+blockysize/2+clamp((z%SIZECANVAS-SIZECANVAS/2)*SHIFTFACTOR)), width=2)


    for x in range(SIZECANVAS):
        for y in range(SIZECANVAS):

            if team[x][y] > -1 and vision[x][y]:

                rect = pygame.Rect(x * blockxsize + blockxsize / 3, y * blockysize + blockysize / 3, blockxsize / 3,
                                   blockysize / 3)
                pygame.draw.rect(SCREEN, colors[team[x][y]], rect)
                text = font.render(f'{troups[x][y]}', True, (0, 0, 0))
                textrect = text.get_rect()
                textrect.center = (x * blockxsize + blockxsize/2, y * blockysize + blockysize/2)
                SCREEN.blit(text, textrect)
    shift=0
    for i in range(len(tomove)):
        i-=shift
        step = (datetime.utcnow() - tomove[i][4]).total_seconds() / timetomove
        if selected[tomove[i][1]][tomove[i][0]]:
            pygame.draw.circle(SCREEN,colors[squadra],((tomove[i][0]+(tomove[i][2]-tomove[i][0])*step)*blockxsize+blockxsize/2,(tomove[i][1]+(tomove[i][3]-tomove[i][1])*step)*blockysize+blockysize/2), 20)
            step = (datetime.utcnow() - tomove[i][4]).total_seconds() / timetomove
            text = font.render(f'{tomove[i][5]}', True, (4, 9, 16))
            textrect = text.get_rect()
            textrect.center = ((tomove[i][0]+(tomove[i][2]-tomove[i][0])*step)*blockxsize+blockxsize/2,(tomove[i][1]+(tomove[i][3]-tomove[i][1])*step)*blockysize+blockysize/2)
            SCREEN.blit(text, textrect)
        if step >= 0.98:

            if tomove[i][6]==tomove[i][7]:
                troups[tomove[i][2]][tomove[i][3]] += tomove[i][5]



            else:

                troups[tomove[i][2]][tomove[i][3]]-=tomove[i][5]
                if troups[tomove[i][2]][tomove[i][3]]<0:
                    vision[tomove[i][2]][tomove[i][3]]+=1
                    neighvisible(tomove[i][2], tomove[i][3], 1)
                    troups[tomove[i][2]][tomove[i][3]]*=-1

                    team[tomove[i][2]][tomove[i][3]]=tomove[i][6]

            sendtroups(tomove[i][2], tomove[i][3], troups[tomove[i][2]][tomove[i][3]])
            sendteam(tomove[i][2], tomove[i][3], team[tomove[i][2]][tomove[i][3]])
            tomove.pop(i)
            shift += 1

    for event in pygame.event.get():
        if event.type == pygame.KEYDOWN:
            if pygame.K_SPACE:
                destroy=1-destroy
        if event.type == pygame.MOUSEBUTTONDOWN:

            if event.button == 4 or event.button == 5:
                if event.button == 4 and xpos<WINDOW_WIDTH:
                    xpos+=WINDOW_WIDTH/20
                elif event.button == 5 and xpos>0:
                    xpos-=WINDOW_WIDTH/20
            else:
                a = int(a // blockxsize)
                b = int(b  // blockysize)

                if a < SIZECANVAS and b < SIZECANVAS and a >= 0 and b >= 0 :
                    if event.button == 3 and vision[a][b]:



                        selected[b][a]=1- selected[b][a]
                        print(a,b)
                        print(selected)
                    elif event.button == 1:
                        if firstsecond==0:
                            print("first: ",a,b)
                            firstsecond=a*SIZECANVAS+b+1
                            print(firstsecond)
                            if team[a][b]!=squadra:
                                firstsecond=-1
                        elif firstsecond!=0 :
                            firstsecond -= 2
                            print("second: ",a,b)
                            print(firstsecond%SIZECANVAS)
                            print(firstsecond//SIZECANVAS)
                            print(15%4)
                            print(15//4)
                            x=firstsecond//SIZECANVAS
                            y=firstsecond%SIZECANVAS
                            print("first connected to second:", connected(x,y,a,b))
                            if (a!=x or b!=y) and quantity>0:
                                if destroy and quantity>=2*troups[a][b]:

                                    troups[x][y]=0
                                    map[y+x*SIZECANVAS][a+b*SIZECANVAS]=0
                                    vision[a][b]-=1

                                else:
                                    if connected(x,y,a,b) :
                                        tomove.append([x,y,a,b,datetime.utcnow(),quantity,team[firstsecond//SIZECANVAS][firstsecond%SIZECANVAS],team[a,b]])
                                        troups[x][y]-=quantity
                                    elif quantity>=2*troups[a][b]:
                                        troups[x][y]=0
                                        map[y+x*SIZECANVAS][a+b*SIZECANVAS]=1
                                        vision[a][b]+=1
                                        pass
                                sendmap(y + x * SIZECANVAS, a + b * SIZECANVAS,
                                        map[y + x * SIZECANVAS][a + b * SIZECANVAS])
                                sendtroups(x, y, troups[x][y])
                            firstsecond=-1


                        firstsecond+=1




        elif event.type == pygame.VIDEORESIZE:
            # There's some code to add back window content here.
            WINDOW_WIDTH,WINDOW_HEIGHT=event.size
            blockxsize = WINDOW_WIDTH / SIZECANVAS
            blockysize = (WINDOW_HEIGHT*6/7) / SIZECANVAS
        elif event.type == pygame.QUIT:
            pygame.quit()

            #WINDOW_WIDTH,WINDOW_HEIGHT=screen.get_width(), screen.get_height()
    CLOCK.tick(30)
    pygame.display.update()


```
